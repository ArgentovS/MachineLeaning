В тестовом датасете нет пропусков. Примем его зрачения правильными и не требующими корректировок.  
  
В тренировочном датасете:
- Пропуски есть в данных признаков f2 (1% пропусков), f7 (25% пропусков), f11 (0,1% пропусков).
 - Устранение пропусков в признаке f2 произведём заменой пропусков на медианное значение, так как большенство значений лежит в 1 квартиле, а разброс значений признака достаточно высок.  
 - Устранение пропусков в признаке f7 произведём заменой пропусков на среднее значение, так как значения возрастают относительно монотонно с каждым квартилем.  
 - Устранение пропусков в признаке f11 произведём удалением строк с пропусками, так как, с одной стороны, строк с пропусками в этом признаке очень мало, а занчит их удаление не исказит тренировочные данные, и, с ругой стороны, распределение признака порказывает что большенство его значений равны 0, а значит среднее или медиана вряд ли могут быть алгоритмами добавляющими достоверность при замене пропусков.  
- Поскольку все признаки датасета являются числовыми - установить ошибки в них можно анализируя данные на аномалии или устанавливая очень сильную корреляцию задублированных признаков. Аномалии выявим тремя методами: попризнаковым, LocalOutlierFactor и IsolationForest  

```python
# Устраним пропуски в признаках f2, f7 и f11
Xtrain['f2'].fillna(Xtrain['f2'].mode().values[0], inplace=True)
Xtrain['f7'].fillna(Xtrain['f7'].mean(), inplace=True)
Xtrain.dropna(subset=['f11'], inplace=True)
Xtrain.info()
```

```python
# Устраним записи-аномалии

# Определим колонки для попризнакового выявления аномалий
check_columns = list(Xtrain.columns)
check_columns.remove('target')

# Закодируем каждый признак, так как метод квантилей работает только с целыми числами
df = Xtrain[check_columns].copy()
for column in check_columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(Xtrain[column])

# Инициализируем словарь индексов, в котором будем фиксировать:
# метод выявления аномалий и соответствующий список индексов записей-аномалий
indexs = {'Quantile': []}

# Заполним словарь индексов записей-аномалий, индексами аномалий,
# выявленными квантилями более 3-сигма
for column in df.columns:
    quantile_indexs = list(df[df[column] > df[column].quantile(0.997)].index)
    if len(quantile_indexs) > 0:
        indexs['Quantile'].extend(quantile_indexs)
indexs['Quantile'] = list(set(indexs['Quantile']))

# Заполним словарь индексов записей-аномалий, индексами аномалий,
# выявленными Локалфактором и Изолесом
lof = LocalOutlierFactor(n_neighbors=100, n_jobs=-1)
ifo = IsolationForest(max_samples=1000, n_jobs=-1, random_state=42)
pred_lof = lof.fit_predict(df)
pred_ifo = ifo.fit_predict(df)
indexs['LocalOutlierFactor'] = list(df[pred_lof == -1].index)
indexs['IsolationForest'] = list(df[pred_ifo == -1].index)

# Сформируем единый список индексов, полученных всеми методами
indexs_all = []
# Определим долю выбросов, выявленную каждым методом
for method in indexs.keys():
    count_anomaly = len(indexs[method])
    print(f"Методом '{method}' выявлено {count_anomaly} записей-аномалий,", end=' ')
    print(f"что составляет {round(count_anomaly/len(Xtrain) * 100, 2)}% всех записей")
    indexs_all.extend(indexs[method])
indexs_all = list(set(indexs_all))
indexs_all.sort()
print(f'\nИТОГО вывленных и удалённых записей-аномалий {len(indexs_all)},', end=' ')
print(f'что составляет {round(len(indexs_all)/len(Xtrain) * 100, 2)}% всех записей')
# Удалим из тренировочного датасета записи с признаками аномалий
Xtrain = Xtrain.drop(index=indexs_all)
Xtrain.reset_index(drop=True, inplace=True)
Xtrain.info()
```

### 3. Постройте baseline-модель:
- разбейте TrainData на тренировочные (Train) и тестовые данные (Test); 
- обучите KNN, LogisticRegression и SVC с параметрами по умолчанию на тренировочных данных (Train);
- примените модели на тестовых данных (Test) и вычислите значение метрики f1.

```python
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")

# Ваш код здесь
x_train, x_train_test, y_train, y_train_test = train_test_split(Xtrain.drop(['target'], axis=1),
                                                                Xtrain['target'],
                                                                test_size=0.3,
                                                                random_state=3)
# Стандартизуем данные
st = StandardScaler()
x_train = pd.DataFrame(data=st.fit_transform(x_train), columns=x_train.columns)
x_train_test = pd.DataFrame(data=st.fit_transform(x_train_test), columns=x_train_test.columns)
x_test = pd.DataFrame(data=st.fit_transform(Xtest.drop(['target'], axis=1)),
                      columns=Xtest.drop(['target'], axis=1).columns)
y_test = Xtest['target']

# Обучим модели
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
lr = LogisticRegression()
lr.fit(x_train, y_train)
cvs = SVC()
cvs.fit(x_train, y_train)
# Модели на данных Xtrain и Xtest
print('\nМетрика f1_score для модели на тренинговых данных:')
for model in [knn, lr, cvs]:
    print(f'  - {model}: {f1_score(model.predict(x_train_test), y_train_test)}')
print('\nМетрика f1_score для модели на тестовых данных:')
for model in [knn, lr, cvs]:
    print(f"  - {model}: {f1_score(model.predict(x_test), y_test)}")
```

```python
# Ваш код здесь
from sklearn.model_selection import GridSearchCV
```

```python
# Модель ближайших соседей
knn_grid = KNeighborsClassifier(n_jobs=-1)
parametrs = {'n_neighbors': [3, 6, 9, 12, 15, 18, 25], #  количество ближайших соседей
             'weights': ['uniform', 'distance'],   #  параметр, не учитывающий / учитывающий удаление соседей
             'p': [1, 2]                           #  параметр мощности
            }
grid = GridSearchCV(knn_grid, parametrs, cv=5)
grid.fit(x_train, y_train)
print(f'Лучшие параметры {knn_grid}:\n{grid.best_params_}')
knn_grid = KNeighborsClassifier(n_jobs=-1,
                                n_neighbors=grid.best_params_['n_neighbors'],
                                p=grid.best_params_['p'],
                                weights=grid.best_params_['weights'])
```

```python
# Модель логистической регрессии
lr_grid = LogisticRegression(n_jobs=-1, class_weight='balanced')
parametrs = {'C': [elem/20 for elem in range(1, 20, 2)],  #  регуляризация (штраф)
             'fit_intercept': [True, False],              #  указание на необходимость добавления константы в функцию потерь
             'solver': ['lbfgs', 'liblinear', 'sag']      # алгоритмы регрессии 
            }
grid = GridSearchCV(lr_grid, parametrs, cv=5)
grid.fit(x_train, y_train)
print(f'Лучшие параметры {lr_grid}:\n{grid.best_params_}')
lr_grid = LogisticRegression(n_jobs=-1,
                             class_weight='balanced',
                             C=grid.best_params_['C'],
                             fit_intercept=grid.best_params_['fit_intercept'],
                             solver=grid.best_params_['solver'])
```

```python
# Модель опорных векторов
svc_grid = SVC(class_weight='balanced', probability=True)
parametrs = {'C': [1, 10, 20],                            #  регуляризация (штраф)
             'kernel': ['linear', 'poly', 'rbf']      #  функция ядра, увеличивающая размерность данных для обеспечения их разделения гиперплоскостью
            }
grid = GridSearchCV(svc_grid, parametrs, cv=5)
grid.fit(x_train, y_train)
print(f'Лучшие параметры {svc_grid}:\n{grid.best_params_}')
svc_grid = SVC(class_weight='balanced', probability=True,
               C=grid.best_params_['C'],
               kernel=grid.best_params_['kernel'])
```

### 5. Оцените на отложенной выборке качество наилучшей модели

### 6. Выполните хитрый трюк
Часто смешивание различных моделей даёт улучшение итогового предсказания. Попробуйте смешать две лучшие модели по формуле:
$$pred_{final} = \alpha\cdot pred_1 + (1-\alpha)\cdot pred_2$$.

Значение $\alpha$ подберите в цикле по Test-выборке. Оцените качество на отложенной выборке.

Удалось ли добиться улучшения качества?

```python
# Ваш код здесь

# Для оценки результата смешивания предсказаний 2-ух лучших моделей
# выберем модели Логистической регрессии и Опрпоных векторов
def mixing_models(x, y, best_score_test):
    best_score_mix = 0
    best_threshold = 0
    best_alpha = 0
    pred_1 = lr_grid.predict_proba(x)[:, 1]
    pred_2 = svc_grid.predict_proba(x)[:, 1]
    for threshold in range(10, 80):
        for alpha in range(1, 101):
            pred = pd.DataFrame(data=(pred_1 * alpha/100 + pred_2 * (1-alpha/100)), columns=['pred_1_2'])
            pred['y_predict'] = pred['pred_1_2'].apply(lambda x: 1 if x > threshold/100 else 0)
            score = f1_score(pred['y_predict'], y)
            if score > best_score_mix:
                best_score_mix = score
                best_threshold = threshold
                best_alpha = alpha
    return f'\nПри смешивании предсказаний двух лучших моделей с коэффициентом {best_alpha/100} '\
           f'и порогом вероятности: {best_threshold/100}\n'\
           f'получился наилучший результат по метрике f1_score: {best_score_mix},\n'\
           f'что на {round((best_score_mix - best_score_test)/best_score_test*100, 2)}% лучше, '\
           f'чем при использовании предсказаний по одной модели'

print(f'\nДля Test-данных: {mixing_models(x_train_test, y_train_test, best_score_test)}')
print(f'\nДля данных отложенной выборки: {mixing_models(x_test, y_test, best_score_test)}')
```
